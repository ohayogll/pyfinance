```python
import requests
payload={
'input_date':'105/05/13',
'select2':'ALL',
'order':'STKNO'  
}
url = 'http://www.twse.com.tw/ch/trading/exchange/BWIBBU/BWIBBU_d.php'
res = requests.post(url, data=payload)
print res.encoding
res.encoding = 'big5'
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(res.text)
tb = soup.select('div#tbl-containerx')[0]
```

```python
import pandas as pd
dfs = pd.read_html(tb.prettify('utf-8'), encoding='utf-8',header = 0, skiprows = 0)
#?pd.read_html  

stockdf = dfs[0]
stockdf.columns = ['id', 'name', 'PER', 'EY', 'PBR' ]
stockdf

#dropna drop NaN data.
```
