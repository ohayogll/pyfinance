```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
stockid = 2332
url = 'http://www.esunsec.com.tw/z/zc/zcp/zcpb/zcpb.djhtm?A={}'
url.format(stockid)
res = requests.get(url)
res.encoding = 'big5'
soup = bs(res.text)
tb = soup.select('table table table')
df = pd.read_html(tb[0].prettify('utf-8'),encoding = 'utf-8')
df[0]
```
