```python
import requests
url = 'http://mops.twse.com.tw/server-java/t164sb01'
payload = {
'step':'1',
'CO_ID':'2059',
'SYEAR':'2013',
'SSEASON':'1',
'REPORT_ID':'C'
}
res = requests.post(url, data = payload)
print res.encoding  # show encoding ISO-8859-1 default encode
```

```python
res.encoding = 'big5'
print res.encoding
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(res.text)
tb = soup.select('table.result_table')[0]
print type(tb)
```

```python
import pandas as pd
dfs = pd.read_html(tb.prettify('utf-8'), encoding = 'utf-8', header = 0, skiprows=0)
dfs[0]
```
